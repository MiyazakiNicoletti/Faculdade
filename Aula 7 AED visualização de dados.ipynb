Gráfico de Barras
O gráfico de barras é utilizado para apresentar dados nominais ou dados numéricos bem segmentados entre diferentes categorias, deixando claramente perceptíveis os resultados comparativos e as tendências nos dados com barras retangulares com alturas e comprimentos proporcionais aos valores que estão sendo representados (FERREIRA; MIRANDA; PINTO, 2021).

Este tipo de gráfico mostra comparações entre dados de várias categorias. Esses gráficos podem ser apresentados em diferentes orientações (vertical ou horizontal) onde um eixo do gráfico mostra as categorias específicas que estão sendo comparadas e o outro eixo representa um valor medido. Há gráficos de barras verticais empilhadas, gráficos de barras horizontais e o mais comum o de barras verticais ou gráficos de colunas.

A biblioteca Matplotlib fornece a função bar() que pode ser usada seguindo a seguinte sintaxe: x.bar(x, height, width, bottom, align). Os parâmetros para a função podem ser definidos como:

x – representa a sequência de dados contendo as coordenadas x das barras. align irá controlar se x é o centro da barra (padrão) ou a borda esquerda.

height(altura) – sequência de escalas que representa a(s) altura(s) das barras.

width(largura) – array, opcional que representa a(s) largura(s) das barras.

bottom(fundo) – array, opcional que representa a(s) coordenada(s) y das barras.

align(alinhar) – Define o alinhamento: {‘center’, ‘edge’}, optional, default ‘center’.

Esta função retorna um objeto container Matplotlib com todas as barras. A seguir um exemplo do gráfico de barras do Matplotlib. Ele mostra o número de alunos matriculados em vários cursos de programação e as saídas são mostradas nas figuras 7.1 e 7.2.

```python
import matplotlib.pyplot as plt
```

```python
linguagens = ['C', 'C++', 'Java', 'Python', 'PHP']

estudantes = [23,17,35,29,12]

plt.bar(linguagens, estudantes)

plt.xlabel("estudantes")

plt.ylabel("linguagens")

plt.title("Número de alunos matriculados por curso ")

plt.show()
```

A função barh() plota o gráfico de barra horizontal como o exemplo a seguir:

```python
linguagens = ['C', 'C++', 'Java', 'Python', 'PHP']

estudantes = [23,17,35,29,12]

plt.barh(linguagens, estudantes)

plt.xlabel("estudantes")

plt.ylabel("linguagens")

plt.title("Número de alunos matriculados por curso ")

plt.show()
```

Histogramas
Forma de representação que mede a distribuição das frequências em relação a determinados intervalos ou faixas. Construído a partir da tabela de frequências, sendo uma ótima ferramenta para avaliar rapidamente uma distribuição de probabilidade que é intuitivamente compreendida.

Por meio do histograma é possível observar os picos, ou seja, os agrupamentos mais altos de barras que representam os valores mais comuns, avaliar a dispersão da amostra para entender o quanto os dados variam, e ainda identificar os valores que estão distantes dos outros valores dos dados (conhecidos como outliers), que podem afetar fortemente os resultados das análises (FERREIRA; MIRANDA; PINTO, 2021).

Em resumo, com o Histograma podemos responder diversas questões sobre:

Centralidade: Onde é esperado que esteja a maioria das observações;

Amplitude: A distribuição contém observações entre quais valores?

Simetria: A distribuição é simétrica ou valores mais altos são mais raros.

Em relação a simetria, podemos avaliar se a distribuição é assimétrica (positiva ou negativa) ou simétrica. No caso da simétrica, a média, a moda e a mediana são os mesmos valores. Na assimetria positiva a Média é maior que a moda e a mediana, já na assimetria negativa a média é menor que a moda e a mediana, conforme pode ser visto na figura 7.3.

```python
import pandas as pd #importa a biblioteca Pandas

import seaborn as srn #importa a Seaborn

df = pd.read_csv('penguins.csv') #Cria o DataFrame da base

srn.histplot(df.loc[:,'bill_depth_mm'], bins = 10, color = 'blue')
```

Boxplot
O boxplot é uma representação gráfica que permite uma análise das distribuições dos dados, baseado no resumo de cinco valores: o valor mínimo, o primeiro quartil, a mediana, o terceiro quartil e o valor máximo. É usado, principalmente, para indicar se a distribuição está distorcida ou não e se há observações incomuns em potencial (outliers) (FERREIRA; MIRANDA; PINTO, 2021).

A figura 7.5 mostra um boxplot apresentando os valores extremos (mínimo e máximo), a mediana e o intervalo interquartil (Q3-Q1). A mediana apresenta uma linha azul e o box um tom de rosa.

Em resumo, os boxplots detalham o valor mínimo, Q1, a mediana, Q3 e o valor máximo de maneira visual. “Os boxplots são úteis, pois fornecem um resumo visual dos dados, permitindo que os pesquisadores identifiquem rapidamente valores médios, a dispersão do conjunto de dados e sinais de assimetria” (FERREIRA; MIRANDA; PINTO, 2021, p. 83).

A distribuição é simétrica quando a mediana está no meio da caixa e os fios conhecidos como “fio de bigode” ou “whiskers” são do mesmo tamanho dos dois lados da caixa, porém, se a mediana está mais próxima do lado esquerda do box e o whisker é mais curto à esquerda da caixa, a distribuição tem assimetria positiva. E, a assimetria negativa é o inverso.

Agora vamos plotar o boxplot da variável ‘df.body_mass_g’ utilizando a biblioteca Seaborn

```python
srn.boxplot(df.body_mass_g).set_title('Peso')
```
