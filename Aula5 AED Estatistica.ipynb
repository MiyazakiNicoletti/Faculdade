Essa base apresenta dados de pinguins de três espécies. Vamos verificar quantos pinguins foram observados em cada espécie e qual a sua proporção em porcentagem?

Para isso, podemos montar uma pequena tabela de frequência. Primeiro precisa-se contar quantos pinguins, depois calcular a porcentagem com o parâmetro “normalize” definido como “True” para retornar a frequência relativa dividindo todos os valores pela soma dos valores, e por fim, montar a tabela (figura 5.2).

```python
frequencia=base['species'].value_counts()

frequencia
```

normalize=True: Este parâmetro, quando definido como True, normaliza os resultados, ou seja, retorna as frequências relativas (proporções) em vez das contagens absolutas. Cada valor único na coluna 'species' terá uma porcentagem associada à sua ocorrência em relação ao total de observações.

```python
porcentagem = base['species'].value_counts(normalize = True )*100

porcentagem
```

```python
distr_frequencia = pd.DataFrame({'Frequência': frequencia, 'Porcentagem (%)': porcentagem})

print(distr_frequencia)
```

Calcular a média
A média aritmética, ou simplesmente média, é a medida de tendência central mais conhecida e utilizada para resumir a informação contida em um conjunto de dados (VIEIRA, 2017).

A média de um conjunto de dados é obtida somando todos os dados e dividindo o resultado pela quantidade.

Média é igual a Soma de todos os dados dividido pelo Total (quantidade). Ou média de x igual a somatória de x dividido por n.
A média é igual ao somatório de x, dividido por n.

Agora, vamos descobrir o peso médio dos pinguins e para isso precisamos indicar a coluna. Para facilitar utilizaremos a biblioteca Pandas.

```python
media = base['body_mass_g'].mean()

media
```

Observe que a média foi calculada considerando 342 e não 344 pois o Pandas por padrão exclui os valores nulos ao calcular o resultado.

Também, podemos observar a média de pesos considerando o sexo dos pinguins utilizando por exemplo a função Groupby()

```python
media = base.groupby(['sex'])['body_mass_g'].mean()

media
```

```python
media = base.groupby(['species'])['body_mass_g'].mean()

media
```

Descobrir a mediana
A mediana divide um conjunto de dados ordenados em duas metades, ou seja, é o valor que ocupa a posição central do conjunto de dados ordenados (VIEIRA, 2017).

Para calcular a mediana dos pesos dos pinguins utilizaremos a função median() do Pandas

```python
mediana = base['body_mass_g'].median()

mediana
```

Isso quer dizer que o valor 4500 ocupa a posição central quando os valores dessa variável estão ordenados. Isso quer dizer que metade dos dados nesta variável estão acima ou abaixo desse valor e podemos comparar esse valor com a média calculada anteriormente observando que possivelmente há valores extremos que estão “puxando” a média.

Observar os quartis
Um conjunto de dados pode ser dividido em mais vezes, não somente em duas como no caso da mediana.

Os quartis dividem um conjunto de dados em quatro partes iguais. O segundo quartil é o mesmo que a mediana e o quarto quartil refere-se ao valor máximo.

Após encontrar os valores dos quartis pode-se calcular a distância interquartil: Terceiro quartil – Primeiro Quartil.

Continuando a observação da variável “body_mass_g” os quartis calculados foram:

Primeiro quartil: Para encontrar o primeiro quartil deve-se informar 25%, ou seja, 0.25.

```python
quartis = base['body_mass_g'].quantile(q=0.25)

quartis
```

Segundo quartil: Deve-se informar 50%, ou seja, 0.50 e assim, encontramos o mesmo valor ao calcular a mediana.

```python
quartis = base['body_mass_g'].quantile(q=0.5)

quartis
```

Terceiro quartil: Deve-se informar 75%, ou seja, 0.75

```python
quartis = base['body_mass_g'].quantile(q=0.75)

quartis
```

Quarto quartil: Informe 100%, ou seja, 1,00 e assim, encontramos o valor máximo que já tínhamos calculado anteriormente.

```python
quartis = base['body_mass_g'].quantile(q=1.00)

quartis
```

Intervalo interquartil
O intervalo interquartil é a diferença entre o terceiro e o primeiro quartis. Assim, considera somente os 50% dos dados e o resultado não é influenciado pelos valores discrepantes, pois os valores discrepantes estão localizados nos 25% de cada lado da distribuição.

Observando a variável “body_mass_g” a distância interquartil é:

Terceiro quartil – Primeiro Quartil:

```python
4750 - 3550
```

Desvio médio absoluto
O termo “desvio” trata-se da diferença absoluta entre um valor e a média dos dados, ou seja, estamos interessados na distância de um valor em relação à média.

Como devemos considerar o módulo (valor absoluto) dos desvios, evitando, deste modo, valores negativos de alguns desvios. Portanto, o desvio médio absoluto será́ dado por:

DM igual a somatória (sendo i igual a um), abre barra vertical - xi menos a média de x - fecha barra vertical, dividido por n.


A biblioteca Pandas fornece um método para tornar esse cálculo mais fácil, mad(), esse método exclui os valores ausentes ao calcular o resultado.

O desvio médio da variável “body_mass_g” é:

```python
base.body_mass_g.mad()
```

Assim, a distância absoluta entre os valores e a média é aproximadamente 676,39. Podemos comparar a dispersão dos dados entre variáveis, considerando que quanto maior o valor do desvio maior a dispersão dos dados.

Um dos problemas de calcular o desvio médio é a dificuldade matemática que o módulo apresenta, claro a biblioteca do Python nos ajuda muito nesse quesito, mas há outra maneira de contornar esse pequeno problema no qual todos os valores são “elevados” ao quadrado assim, calculamos a média dos quadrados dos desvios que chamamos de variância.

Variância
A variância informa como os valores de uma variável se distribuem em torno da média. Assim, quanto maior a variância, mais longe da média estarão alguns valores.

```python
base.body_mass_g.var()
```

Desvio Padrão
Geralmente os dados têm unidades de medidas como quilogramas, metros e entre outros e ao elevar os desvios ao quadrado essas unidades também são calculadas ao quadrado assim, dificultando a comparação do valor da variância com os dados. Para resolver isso, podemos extrair a raiz quadrada do resultado da variância que é chamado desvio padrão. Quando o desvio padrão está alto demais em relação às medidas de centralidade (média e mediana) sugere que há valores extremos incomuns que estão afetando a distribuição de valores em torno da média (AKANIME; YAMAMOTO, 2013).

O desvio padrão populacional é a raiz quadrada da variância populacional, ou seja:

Sigma igual a raiz de sigma ao quadrado.
E, o desvio padrão amostral (s) é a raiz quadrada da variância amostral, ou seja:

s igual a raiz de s ao quadrado.
A biblioteca Pandas, também, fornece um método que retorna o desvio padrão o std(), assim o desvio padrão calculada da variável “body_mass_g” é:

```python
base.body_mass_g.std()
```

Coeficiente de Variação
Quando precisamos comparar duas distribuições, por exemplo, não podemos fazer uma comparação direta, uma vez que a magnitude dos dados dessas distribuições pode ser muito diferente. Podemos calcular o coeficiente de variação que é uma medida que expressa a variabilidade dos dados, mas exclui a influência da ordem de grandeza.

O coeficiente de variação analisa a dispersão em termos relativos em porcentagem (%). Quanto menor for o seu valor mais homogêneo serão os dados, ou seja, menor será a dispersão em torno da média. Para calcular divide-se o desvio padrão pela média:

cv igual a sigma dividido por u ou cv igual a s dividido por x.
Podemos calcular o coeficiente de variação da variável “body_mass_g” com o seguinte código:

```python
coefVar = (base.body_mass_g.std())/(base['body_mass_g'].mean())*100

print(f' Coeficiente de variação: {coefVar:.2f} %')
```
