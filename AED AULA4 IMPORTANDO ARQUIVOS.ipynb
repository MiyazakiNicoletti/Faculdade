Agora vamos entender as formas de agrupar os dados carregados. Para isso utilizaremos a função Groupby criando nossos agrupamentos de categorias. Este é um conceito extremamente simples, mas é uma técnica muito valiosa, amplamente usada na análise de dados. A função groupby( ) é realmente importante por causa de sua capacidade de resumir, agregar e agrupar dados com eficiência.

Para descobrirmos quais serão os agrupamentos que serão criados, por exemplo, com a característica “Espécies” da base Iris, utilizaremos o método unique() que é usado para obter todos os valores únicos de uma determinada coluna.
Acrescentando o método unique(), sua execução nos mostrará um array com os valores únicos da coluna espécies.

```python
df['Species'].unique()
```

Se utilizarmos a função nunique() ela retornará o valor “3”, pois esta função nos mostra a quantidade dos valores únicos da coluna especificada. Podemos usar também:

```python
df['Species'].nunique()
```

```python
df.info()
```

Agora que sabemos as informações contidas em nosso DataFrame, podemos iniciar alguns agrupamentos, e para isso vamos digitar o código a seguir:

```python
g = df.groupby('Species')
```

```python
g.head()
```

Podemos aproveitar o agrupamento e utilizar as funções min(), max(), count(), sum() e mean(), tendo assim respectivamente o mínimo, máximo, total, soma e média das informações agrupadas. Veja o exemplo a seguir:

```python
g = df.groupby('Species').min()
g.head()
```

```python
g = df.groupby('Species').max()
g.head()
```

```python
g = df.groupby('Species').count()
g.head()
```

```python
g = df.groupby('Species').sum()#exibe a soma dos valores

g.head()
```

```python
g = df.groupby('Species').mean()#exibe a média dos valores

g.head()
```

utilizamos o método query() passando como parâmetro a coluna Sepal.Length onde a condição de retorno é que seus valores sejam maiores que 5.7. Podemos ver o resultado na figura 4.14.

```python
df.query('`Sepal.Length` > 5.7')
```

IMPORTANTE: Para colunas com espaços ou caracteres especiais em seu nome, você pode usar aspas com acento grave: df.query('`Sepal.Length` > 5.7')
